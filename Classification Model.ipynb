# Cleaning Data

```python
df=df.drop(["Email Address"],axis=1)
df=df.drop(["First Name"],axis=1)
df=df.drop(["Last Name"],axis=1)
df=df.drop(["Zip Code"],axis=1)
df=df.drop(["Contact Number"],axis=1)
df=df.drop(["Emergency Contact Number"],axis=1)
df=df.drop(["University Name"],axis=1)
df=df.drop(["Degree"],axis=1)
df=df.drop(["Course Type"],axis=1)
df=df.drop(["Current Employment Status"],axis=1)
df=df.drop(["Expected Graduation-year"],axis=1)
df=df.drop(["State"],axis=1)
df=df.drop(["Certifications/Achievement/ Research papers"],axis=1)
df=df.drop(["DOB [DD/MM/YYYY]"],axis=1)
df=df.drop(["Link to updated Resume (Google/ One Drive link preferred)"],axis=1)
df=df.drop(["link to Linkedin profile"],axis=1)
```

```python
# Check the data agian for null values
df.isnull().sum()
```

# Encoding Label

```python
lbl=df["Label"].tolist()
x=[]
for i in range(0,len(lbl)):
    if lbl[i]=='eligible':
        x.append(1)
    else:
        x.append(0)
```

```python
Label=pd.DataFrame({"Label":x})
df=df.drop(["Label"],axis=1)
df=pd.concat([df,Label],axis=1,sort=False)
```

# Preprocessing and Splitting data

```python
from sklearn import preprocessing
X = np.asarray(dcode[['City', 'Age', 'Gender', 'College name', 'Major/Area of Study', 'Which-year are you studying in?', 'CGPA/ percentage','Areas of interest','Have you worked core Java','Programming Language Known other than Java (one major)','Have you worked on MySQL or Oracle database','Have you studied OOP Concepts','Rate your written communication skills [1-10]','Rate your verbal communication skills [1-10]','How Did You Hear About This Internship?']])
y = np.asarray(dcode['Label'])
X = preprocessing.StandardScaler().fit(X).transform(X)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=0.2, random_state=4)
```

# SVM for calculating F1 score

```python
from sklearn import svm
from sklearn.svm import SVC
from sklearn.metrics import jaccard_similarity_score

clf1 = svm.SVC(kernel='rbf')
clf1.fit(X_train, y_train) 
yhat2 = clf1.predict(X_test)
#jaccard_similarity_score(y_test, yhat2)
model = SVC(gamma='scale')
```

```python
from sklearn.metrics import f1_score
print("F1 Score is :",f1_score(y_test,yhat2))
```
